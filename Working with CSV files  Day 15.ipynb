## Opening the CSV file from an URL:

```python
import requests
from io import StringIO

url = " https://www.kaggle.com/datasets/reddynitin/aug-train"
headers={"User-Agent":"Mozilla/5.0 (Macintosh;intel Mac Os X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req=requests.get(url, headers=headers)
data = StringIO(req.txt)

pd.read_csv(data)
```

## Converter:

```python
pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
def rename(name):
    if name=="Royel Challengers Bangalor":
        return "RCB"
    else:
        return name
```

```python
rename("Royel Challengers Bangalor")
```

```python
pd.read_csv('IPL Ball-by-Ball 2008-2020.csv',converters={'batting_team':rename})
```
